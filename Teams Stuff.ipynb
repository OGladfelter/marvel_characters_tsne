```python
import pandas as pd
import re
```

```python
data = pd.read_csv("Top_Marvel_characters_With_Metadata.csv")
```

## First, we clean

```python
def cleanTeamsList(teamsList):
    messy_teams_list = teamsList.split(",")
    clean_teams_list = []
    
    for team in messy_teams_list:
        cleaned_team = re.sub("[\(\[].*?[\)\]]", '', team) # remove info in parentheses
        cleaned_team = cleaned_team.strip(" ")
        
        # ALSO NEED TO REMOVE 'former' HERE
        
        clean_teams_list.append(cleaned_team)
        
    return clean_teams_list
```

```python
data['teams_clean'] = data['teams'].apply(cleanTeamsList)
```

```python
data.to_csv("Top_Marvel_characters_With_Metadata.csv")
```

## Second, we count

```python
teams_count = {}

for num in range(0,len(data)):
    if data['teams'][num] == -1:
        continue
        
    teams = data['teams_clean'][num]
    
    for team in teams:
        team = team.strip(" ")
        if team in teams_count:
            teams_count[team] = teams_count[team] + 1
        else:
            teams_count[team] = 1
```

```python
teams_count_df = pd.Series(teams_count, name='value')
teams_count_df.index.name = 'Team'
teams_count_df = teams_count_df.reset_index()
```

```python
teams_count_df = teams_count_df[teams_count_df['value'] > 11] # why 11? because sinister six has 12 hehe
```

```python
teams_count_df.sort_values('value', ascending=False).head()
```

```python
teams_count_df = teams_count_df.reset_index()
del teams_count_df['index']
```

## Third, we munge
I need a data set with the following columns: “team”, “members”, “member_ids”. 

One row would be: “Avengers”, “Captain America, Iron Man, Spiderman, Thor”, “1,2,0,4”

```python
team_members = []
team_members_ids = []

for num in range(0,len(teams_count_df)):
    names = []
    ids = []
    
    for i in range(0,len(data)):
        for team in data['teams_clean'][i]:
            if team == teams_count_df['Team'][num]:
                names.append(data['nickName'][i])
                ids.append(data['pic_id'][i])
                
    team_members.append(names)
    team_members_ids.append(ids)
```

```python
len(team_members) == len(team_members_ids) == len(teams_count_df)
```

```python
teams_count_df['team_members'] = team_members
teams_count_df['team_members_ids'] = team_members_ids
```
